# Ultra-Realistic Alien World Generator

This notebook guides you through the process of converting a rough sketch into a photorealistic 3D alien planet.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import Image, Video, display
import os

import wp1_fractal_coastline as wp1
import wp2_heightmap_synthesis as wp2
import wp3_biome_texturing as wp3
import wp4_cloud_layer as wp4
import wp5_final_renderer as wp5
import weorold_ultra as globe

def show_img(path, title=None, width=None):
    if title: print(f'### {title} ###')
    display(Image(filename=path, width=width))

print('Environment Ready.')
```

## WP1: Coastline Fractalization
We take the input sketch and use a Signed Distance Field (SDF) displaced by fractal noise.

```python
INPUT_SKETCH = 'Motoki_Aspsp_uk_Fig02_c_Islands.jpg'
wp1.process_sketch_to_fractal_mask(INPUT_SKETCH, upscale_factor=4, noise_strength=20.0)

show_img('wp1_comparison.png', title='Original Sketch vs Fractal Coastline', width=1000)
```

## WP2: Heightmap Synthesis
Generating a 16-bit elevation map.

```python
wp2.synthesize_heightmap('wp1_fractal_mask.png')

show_img('wp2_height_preview.png', title='Terrain Elevation (Preview)', width=600)
```

## WP3: Biome & Surface Texturing
Assigning realistic colors.

```python
wp3.create_surface_texture('wp2_height_map.png')

show_img('wp3_surface_texture.png', title='Surface Texture (Biomes)', width=800)
```

## WP4: Atmospheric Cloud Layer
Generating procedurally warped clouds.

```python
h_img = cv2.imread('wp2_height_map.png', cv2.IMREAD_UNCHANGED)
wp4.create_cloud_layer(h_img.shape)

show_img('wp4_cloud_map.png', title='Procedural Cloud Map', width=800)
```

## WP5: Final 2D Map Rendering
Combining all layers with Phong shading.

```python
wp5.render_final_maps()

show_img('wp5_shaded_clouds.png', title='Final Shaded Map with Clouds', width=1000)
```

## Final Result: 3D Rotating Globe
Projecting the final textures onto a 3D sphere.

```python
globe.create_rotating_globe('wp5_shaded_clouds.png', 'ultra_globe_rotation.mp4', frames=60, fps=20)

print('\nDisplaying Final Globe Animation:')
display(Video('ultra_globe_rotation.mp4', embed=True, width=600))
```
