# Alien Planet Globe Generator

This notebook generates a high-definition 3D globe animation from a 2D map image using an ultra-smooth OpenCV renderer.

```python
import cv2
import numpy as np
from IPython.display import Video, display
import weorold_ultra as globe

# --- Configuration ---
MAP_IMAGE_PATH = 'Motoki_Aspsp_uk_Fig02_c_Islands.jpg'
OUTPUT_ANIMATION_PATH = 'globe_rotation.mp4'

print("Environment Ready.")
```

## Generate 3D Animation
Click the button below to render the rotating globe using the high-fidelity perspective renderer.

```python
print(f"🎬 Rendering 3D animation from {MAP_IMAGE_PATH}...")
globe.create_rotating_globe(MAP_IMAGE_PATH, OUTPUT_ANIMATION_PATH, frames=60, fps=20)

print('\nDisplaying Final Globe Animation:')
display(Video(OUTPUT_ANIMATION_PATH, embed=True, width=600))
```
